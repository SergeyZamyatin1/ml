```python
from sklearn import linear_model
import csv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
date = []
total_cases = []
new_cases = []

with open("owid-covid-data.csv", "r") as file:
    reader = csv.reader(file, delimiter = ',')
    for row in reader:
        if row[2] == "Russia":
            date.append(row[3])
            total_cases.append(float(row[4]))
            new_cases.append(float(row[5]))

total_cases = np.asarray(total_cases)
new_cases = np.asarray(new_cases)
# Создаём равномерную сетку
date = np.linspace(0, 1, len(date))
```

```python
fig, axes = plt.subplots(2)
fig.suptitle('Все случаи')
axes[0].plot(date, total_cases)
axes[1].plot(date, np.log(total_cases), label="Логарифм числа случаев")
axes[1].legend()
```

```python
fig, axes = plt.subplots(2)
fig.suptitle('Новые случаи')
axes[0].plot(date, new_cases)
axes[1].plot(date, np.log(new_cases), label="Логарифм числа новых случаев")
axes[0].legend()
axes[1].legend()
```

```python
# Берем участок с экспоненциальным ростом
x = date[50:80] 
y = total_cases[50:80]
v = new_cases[50:80]
plt.plot(x, y)
```

```python
model = linear_model.LinearRegression(fit_intercept=True).fit(np.asarray([x]).T, np.log(y))
print(model.intercept_)
print(model.coef_)

plt.title("Логарифм числа случаев")
plt.plot(x, np.log(y), label="Реальные данные")
plt.plot(x, model.predict(np.asarray([x]).T), label="Линейная регрессия")
plt.legend()
```

```python
# Дисперсия данных.
dif = np.log(y)-model.predict(np.asarray([x]).T)
D = (np.mean(dif**2)-np.mean(dif)**2)**0.5
# Задаем априорное распределение.
C0 = np.asarray([[1, 0], [0, 1]])*1000
m0 = np.asarray([[0, 0]]).T
# Вычисление параметров апостериорного.
X = np.hstack([np.ones((len(x), 1)), np.asarray([x]).T])
C = np.linalg.inv(np.linalg.inv(C0)+1/D**2*np.matmul(X.T, X))
m = np.matmul(C, np.matmul(np.linalg.inv(C0), m0)+1/D**2*np.matmul(X.T, np.asarray([np.log(y)]).T))

print(C)
print(m)
```

```python
# X - матрица признаков, W - матрица из 100 случайных векторов весов, распределенных 
# по априорному распределению, то есть семплирование
X = np.hstack([np.ones((len(x), 1)), np.asarray([x]).T])
w = np.random.multivariate_normal(m.T[0], C, 100).T
Y = np.matmul(X, w).T
for yi in Y:
    plt.plot(x, yi, linewidth=.1)

plt.title("Логарифм числа случаев")
plt.plot(x, np.log(y), label="Реальные данные")
plt.plot(x, np.mean(Y, axis=0), linewidth=2, label="Среднее")
plt.legend()

#Число заболевших на сегодня:
print(np.exp(model.predict([[1]])))
```

```python
# Берем участок с плато для сигмоиды
x = date[50:150] 
y = total_cases[50:150]
v = new_cases[50:150]

# Матрица признаков, но уже с квадратами данных
X = np.hstack([np.ones((len(x), 1)), np.asarray([x]).T, (np.asarray([x]).T)**2])
# Матрица признаков без единиц
_X = np.hstack([np.asarray([x]).T, (np.asarray([x]).T)**2])
model = linear_model.LinearRegression(fit_intercept=True).fit(_X, np.log(v))
print(model.intercept_)
print(model.coef_)

plt.title("Логарифм числа новых случаев")
plt.plot(x, np.log(v), label="Реальные данные")
plt.plot(x, model.predict(_X), label="Максимальное правдоподобие")
plt.legend()
```

```python
# Дисперсия данных.
dif = np.log(y)-model.predict(_X)
D = (np.mean(dif**2)-np.mean(dif)**2)**0.5
# Задаем априорное распределение.
C0 = np.asarray([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
m0 = np.asarray([[model.intercept_, model.coef_[0], model.coef_[1]]]).T
# Расчет апостериорного
C = np.linalg.inv(np.linalg.inv(C0)+1/D**2*np.matmul(X.T, X))
m = np.matmul(C, np.matmul(np.linalg.inv(C0), m0)+1/D**2*np.matmul(X.T, np.asarray([np.log(v)]).T))
print(C)
print(m)
```

```python
#Семплирование
X = np.hstack([np.ones((len(x), 1)), np.asarray([x]).T, (np.asarray([x]).T)**2])
w = np.random.multivariate_normal(m.T[0], C, 100).T
Y = np.matmul(X, w).T
for yl in Y:
    plt.plot(x, yl, linewidth=.1)

plt.title("Логарифм новых случаев")
plt.plot(x, np.log(v), label="Реальные данные")
plt.plot(x, np.mean(Y, axis=0), linewidth=2, label="Усреднение")
plt.legend()
```

```python
# Суммирование новых случаев по дням
Y = np.exp(Y)

Sum = []
for yi in Y:
  t = 0
  I = []
  for z in yi:
    t += z
    I += [t]
  Sum.append(I)
Sum = np.asarray(Sum)

print(Sum)
for s in Sum:
    plt.plot(x, s, linewidth=.1)

plt.title("Число случаев")
plt.plot(x, y, label="Реальные данные")
plt.plot(x, np.mean(Sum, axis=0), linewidth=2, label="Усреднение")
plt.legend()
```
